```python
bin_path = "3D_data_bin_processed.tif"
gray_path = "3D_data_gray.tif"
```

```python
a_bin = np.array(tifffile.imread(bin_path))
print(a_bin.shape)
a_gray = np.array(tifffile.imread(gray_path))
print(a_gray.shape)
```

```python
def show_random_slices(a):
    figs, axes = [], []
    for _ in range(3):
        fig, ax = plt.subplots()
        figs.append(fig)
        axes.append(ax)
        
    x_max, y_max, z_max = a.shape
    x = np.random.randint(x_max)
    y = np.random.randint(y_max)
    z = np.random.randint(z_max)
    x_slice = a[x,:,:]
    x_slice = a[x,100:200,100:200]
    y_slice = np.rot90(a[:,y,:])
    z_slice = np.rot90(a[:,:,z])
    axes[0].imshow(x_slice, cmap = "gray")
    axes[1].imshow(y_slice, cmap = "gray")
    axes[2].imshow(z_slice, cmap = "gray")
    
    return figs, axes
```

```python
show_random_slices(a_bin)
```
