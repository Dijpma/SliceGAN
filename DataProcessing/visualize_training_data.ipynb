```python
def show_random_slices(a):
    figs, axes = [], []
    for _ in range(3):
        fig, ax = plt.subplots()
        figs.append(fig)
        axes.append(ax)
        
    x_max, y_max, z_max = a.shape
    x = np.random.randint(x_max)
    y = np.random.randint(y_max)
    z = np.random.randint(z_max)
    x_slice = a[x,:,:]
    y_slice = np.rot90(a[:,y,:])
    z_slice = np.rot90(a[:,:,z])
    print(np.unique(a_bin))
    print("Amount of color values: ", np.unique(a_bin).size)
    axes[0].imshow(x_slice, cmap = "gray")
    axes[1].imshow(y_slice, cmap = "gray")
    axes[2].imshow(z_slice, cmap = "gray")
    #
    # tifffile.imwrite("./../TrainingData/3D_data_bin_slice_x.tif", x_slice)
    # tifffile.imwrite("./../TrainingData/3D_data_bin_slice_y.tif", y_slice)
    # tifffile.imwrite("./../TrainingData/3D_data_bin_slice_z.tif", z_slice)
    print("done")
    return figs, axes
```

```python
show_random_slices(a_gray)
```
