```python
def show_random_slices(a):
    figs, axes = [], []
    for _ in range(3):
        fig, ax = plt.subplots()
        figs.append(fig)
        axes.append(ax)
        
    x_max, y_max, z_max = a.shape
    x = np.random.randint(x_max)
    y = np.random.randint(y_max)
    z = np.random.randint(z_max)
    x_slice = a[x,:,:]
    y_slice = np.rot90(a[:,y,:])
    z_slice = np.rot90(a[:,:,z])
    axes[0].imshow(x_slice, cmap = "gray")
    axes[1].imshow(y_slice, cmap = "gray")
    axes[2].imshow(z_slice, cmap = "gray")

    # tifffile.imwrite("./../Training_Data/data_gray_x.tif", x_slice)
    # tifffile.imwrite("./../Training_Data/data_gray_y.tif", y_slice)
    # tifffile.imwrite("./../Training_Data/data_gray_z.tif", z_slice)

    return figs, axes
```

```python
show_random_slices(a_gray + 128)
```
